```python
l=[]
base_url=("https://www.sanasafinaz.com/pk/ready-to-wear.html?p=")
for page in range(1,8):
    print(base_url+str(page)+".html")
    site=requests.get(base_url+str(page)+".html")
    #print(site.status_code)
    soup=BeautifulSoup(site.content, "html.parser")
    #print(soup.prettify())
    all=soup.find_all("div", {"class", "product-item-details"})
    #len(all)
    for item in all:
        d={}
        d["item"]=item.find("a", {"class", "product-item-link"}).text.replace("\n", "").replace(" ", "")
        try:
            d["regular-price"]=item.find_all("span", {"class", "price"})[0].text.replace(" ", "").replace("PKR", "")
        except:
            d["regular-price"]=None
        try:
            d["old-price"]=item.find_all("span", {"class", "price"})[1].text.replace(" ", "").replace("PKR", "")
        except:
            d["old-price"]=None
            
        d["url"]=item.find("a", {"class", "product-item-link"}).get('href')
        #print(item.find("div", {"id", "option-label-size-580-item-40"}).text
        #print("")
        l.append(d)
```
